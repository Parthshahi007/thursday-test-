```python
customer_feedback = """Your recent order experience has been satisfactory overall. While there were some minor issues,
we appreciate the effort made to resolve them promptly."
"""

# List of words to search for
key_comments = {
    0: 'satisfactory',
    1: 'order',
    2: 'effort',
    3: 'issues',
    4: 'promptly',
    5: 'appreciate',
    6: 'experience',
    7: 'resolve',
    8: 'overall',
    9: 'minor'
}

# Write your search code here and provide comments. 

first_digit = 7  
last_digit = 0  

allocated_keys = [first_digit, last_digit]

# List to store start and end positions
my_list = []


for key in allocated_keys:
    word = key_comments[key]
    start_pos = customer_feedback.find(word)
    if start_pos != -1:
        end_pos = start_pos + len(word) 
        my_list.append((start_pos, end_pos))

# Print the result
print(my_list)
```

```python
first_2 = 75  
last_2 = 70 

# calculate Operating Profit Margin
def operating_profit_margin(revenue, operating_income):
    if revenue != 0:
        return (operating_income / revenue) * 100
    else:
        return 0
   
#  calculate Revenue per Customer
def revenue_per_customer(total_revenue, total_customers):
    if total_customers != 0:
        return total_revenue / total_customers
    else:
        return 0


# Function to calculate Customer Churn Rate
def customer_churn_rate(lost_customers, total_customers):
    if total_customers != 0:
        return (lost_customers / total_customers) * 100
    else:
        return 0


# calculate Average Order Value
def average_order_value(total_revenue, total_orders):
    if total_orders != 0:
        return total_revenue / total_orders
    else:
        return 0

# Example calculations using ID digits as input values
revenue = first_2 * 1000   
operating_income = last_2 * 100  
total_customers = first_2 * 10   
lost_customers = last_2  
total_orders = first_2 * 5 

# Calling the functions
opm = operating_profit_margin(revenue, operating_income)
rpc = revenue_per_customer(revenue, total_customers)
ccr = customer_churn_rate(lost_customers, total_customers)
aov = average_order_value(revenue, total_orders)

# Printing the results
print(f"Operating Profit Margin: {opm}%")
print(f"Revenue per Customer: {rpc}")
print(f"Customer Churn Rate: {ccr}%")
print(f"Average Order Value: {aov}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
prices = np.array([20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]).reshape(-1, 1)
demand = np.array([300, 280, 260, 240, 210, 190, 160, 140, 120, 100, 85])

# Create and train the linear regression model
model = LinearRegression()
model.fit(prices, demand)

# Function to predict demand at a given price
def predict_demand(price):
    return model.predict(np.array([[price]]))[0]

# Predict demand when the price is set at £52
demand_at_52 = predict_demand(52)

# Finding the price that maximizes revenue
price_range = np.linspace(20, 70, 100)  # Generate price values
predicted_demand = model.predict(price_range.reshape(-1, 1))
revenue = price_range * predicted_demand
optimal_price = price_range[np.argmax(revenue)]

# Display results
print(f"Predicted demand at £52: {demand_at_52:.2f} units")
print(f"Optimal price for maximum revenue: £{optimal_price:.2f}")

# Plot revenue vs. price
plt.figure(figsize=(8, 5))
plt.plot(price_range, revenue, label="Revenue Curve", color='blue')
plt.axvline(optimal_price, color='red', linestyle='--', label=f'Max Revenue at £{optimal_price:.2f}')
plt.xlabel("Price (£)")
plt.ylabel("Revenue (£)")
plt.title("Revenue vs Price")
plt.legend()
plt.show()
```
